```python
now=datetime.datetime.now().date()
y = now - timedelta(1)

def stock_sentiment(quote, num_tweets):
    list_of_tweets = user.search(quote, count=num_tweets, since=y, until=now)
    print (len(list_of_tweets))
    polarity_list = []
    pos = []
    neg = []
    for tweet in list_of_tweets:
        blob = TextBlob(tweet.text).sentiment
        polarity = blob.polarity
        if polarity > 0:
            pos.append(polarity)
        elif polarity < 0:
            neg.append(polarity)
        polarity_list.append(polarity)
    return sum(polarity_list)/(len(pos)+len(neg))

stock_sentiment('FB', 200)
```
